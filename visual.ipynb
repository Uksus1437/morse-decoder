```python
import torch
import torchaudio
import torchaudio.transforms as T
import matplotlib.pyplot as plt
```

```python
sample_rate = 8000
n_mels = 40
mel = T.MelSpectrogram(sample_rate=sample_rate, n_mels=n_mels)
```

```python
waveform1 = torch.load("D:/vs_projects/Data/morse/morse_wav/3001.pt")
waveform2 = torch.load("D:/vs_projects/Data/morse/morse_wav/2000.pt")
waveform3 = torch.load("D:/vs_projects/Data/morse/morse_wav/31501.pt")
waveform4 = torch.load("D:/vs_projects/Data/morse/morse_wav/32001.pt")
```

```python
spec1 = mel(waveform1).squeeze().log2()
spec2 = mel(waveform2).squeeze().log2()
spec3 = mel(waveform3).squeeze().log2()
spec4 = mel(waveform4).squeeze().log2()
```

```python

plt.figure(figsize=(20, 7))

plt.subplot(2, 2, 1)
plt.title('train')
plt.imshow(spec1.numpy(), aspect='auto', origin='lower', cmap='viridis')
plt.subplot(2, 2, 2)
plt.title('train')
plt.imshow(spec2.numpy(), aspect='auto', origin='lower', cmap='viridis')

plt.subplot(2, 2, 3)
plt.title('test')
plt.imshow(spec3.numpy(), aspect='auto', origin='lower', cmap='viridis')
plt.subplot(2, 2, 4)
plt.title('test')
plt.imshow(spec4.numpy(), aspect='auto', origin='lower', cmap='viridis')
```
